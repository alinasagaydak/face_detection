```python
import cv2
import os
import numpy as np
```

```python
folders = os.listdir("/Users/alina/jupiter/Wider_face_dataset_2/images_dataset")
```

```python
def boxes(path):
    os.chdir("/Users/alina/jupiter/Wider_face_dataset_2")
    txt = open ("wider_face_train_bbx_gt.txt", "r")
    data = txt.read()
    X = []
    lines = data.split('\n')
    for i, line in enumerate(lines):
        if line == path:
            num_of_boxes = int(lines[i+1])
            for j in range(num_of_boxes):
                box = lines[i+2+j].split(' ')[:4]
                box[0] = int(box[0])
                box[1] = int(box[1])
                box[2] = int(box[2]) + box[0]
                box[3] = int(box[3]) + box[1]
                
                X.append(box)
    return X
```

```python
i = 0
X=[]
for folder in folders:
    path = "/Users/alina/jupiter/Wider_face_dataset_2/images_dataset/" + folder
    names = os.listdir(path)
    names = names[:40]
    for name in names:
        X.append(boxes(folder+"/"+name))
        img = cv2.imread(path+"/"+name)
        os.chdir ("/Users/alina/jupiter/Wider_face_dataset_2/wider_face_train")
        cv2.imwrite(f"{i}.jpg", img)
        i += 1
    i = i
```
